## 2. 栈

### 2.1 栈的含义

栈(stack),又称堆栈,是保证元素**后进先出**(LIFO, Last In First Out)关系的线性表.


    1.栈中元素只有一种关系:到来的先后顺序.
    2.栈是一种容器,可存入,访问和删除栈中的数据.
    3.栈顶与栈底,入栈与出栈
    4.只能在栈顶存入,访问和删除数据.
    
 ### 2.2 队列
 队列(queue)是保证元素**先进先出**(First In First Out)关系的线性表.
 
    1.队列中没有位置的概念. 
    2.队列是一种容器,可存入,访问和删除队列中的数据
    3.队头和队尾,入队和出队
    4.队列只支持默认方式的元素存入和取出.
    5.队列是只能在一端存入元素,在另一端删除操作的线性表.(不允许在其他位置进行操作)
    

### 2.3 栈的实现
1.我们可以用python程序来实现栈这一数据结构. 同顺序表一样,我们将栈定义成**类**.

2.栈的操作：

    1. 判断是否为空: is_empty()
    2. 求栈的元素个数: size()
    3. 入栈: push(item)
    4. 出栈 : pop() 
    5. 访问栈顶元素: top()


```python
# 栈的实现
# 实现的方式： 以顺序表为基础
class Stack():
    # 生成一个容器
    def __init__(self):
        self.__list = []
        
    # 判断是否为空
    def is_empty(self):
        return self.__list == []
    
    # 入栈操作： 向栈顶添加元素
    def push(self,item):
        self.__list.append(item)
        
    # 出站操作： 取出栈顶（元素）
    def pop(self):
        # 当栈为空的时候，返回None
        if self.__list == []:
            return None 
        return self.__list.pop()
        
    # 访问栈顶元素： 不会删除元素
    def top(self):
        # 当栈为空的时候，返回None
        if self.__list == []:
            return None
        return self.__list[-1]
    
    # 求栈中的元素的个数
    def size(self):
        return len(self.__list)
        

if __name__ == "__main__":
    st = Stack()
    print(st.is_empty())
    
    #入栈
    st.push(10)
    st.push(20)
    st.push(100)
    st.push(200)
    st.push("agent!")
    
    print(st.is_empty())
    print(st.size())
    print("栈顶元素：",st.top())
    print("出栈：",st.pop())
    print(st.size())
    print("栈顶元素：",st.top())
    
    
    
```

### 2.5 队列的实现和基本操作


```python
# 基于顺序表来实现队列
class Queue():
    """队列"""
    def __init__(self):
        self.__list=[]
        
    #判断队列是否为空
    def is_empty(self):
        return self.__list==[]
    
    # 入队操作：加入一个元素到队列尾
    def enqueue(self, item):
        self.__list.append(item)# 列表头部作为队头 T = O(1)
        
    # 求出队列的元素个数
    def size(self):
        return len(self.__list)
    
    # 访问队列头
    def top(self):
        if self.__list==[]:
            return None
        return self.__list[0]
    
    # 出队操作：把队列头部的元素取出
    def dequeue(self):
        if self.__list==[]:
            return None
        return self.__list.pop(0)
    
if __name__ =="__main__":    
    q = Queue()
    q.enqueue(3)
    q.enqueue(5)
    q.enqueue(30)
    q.enqueue(50)
    print(q.is_empty())
    print("元素个数",q.size())
    print("访问队列头：",q.top())
    print("出队：",q.dequeue())
    print("出队：",q.dequeue())
    
    
    
```

### 2.6 双端队列

特征: 双端队列可在任意一端出队和入队.

基本操作：

1.双端队列的操作:

    创建双端队列
        deque()
    队列是否为空
        is_empty()
        
    从队头加入元素
        add_front(item)
        
    从队尾加入元素
        add_rear(item)
        
    从队头删除一个元素
        rm_front()
        
    从队尾删除一个元素
        rm_rear()
        
    元素个数(返回双端队列的大小)
        size()

```python
class Deque():
    def __init__(self):
        self.__list =[]
    
    #判断队列是否为空
    def is_empty(self):
        return self.__list==[]
    
    #在队头加入元素
    def add_front(self,item):
        # 队头用列表头部表示
        self.__list.insert(0,item) # T =O(N)
    # 在队尾加入元素
    def add_rear(self,item):
        # 队头用列表头部表示
        self.__list.append(item) # T =O(1)
        
    # 在队头出队
    def rm_front(self):
        # 把列表的头部的元素删除. 队头用列表头部表示
        if self.__list ==[]:
            return None
        return self.__list.pop(0)   # T = O(N)
    # 在队尾出队
    def rm_rear(self):
        # 把列表的尾部的元素删除. 队头用列表头部表示
        if self.__list ==[]:
            return None
        return self.__list.pop()  # T =O(1)
    
    # 提取元素个数
    def size(self):
        return len(self.__list)

if __name__ =="__main__":    
    dq = Deque()
    print(dq.is_empty())
    dq.add_front(10)
    dq.add_front(20)
    dq.add_rear(100)
    dq.add_rear(200)
    print(dq.is_empty())
    print(dq.size())
    dq.rm_front()
    dq.rm_rear()
    print(dq.size())
    
```
